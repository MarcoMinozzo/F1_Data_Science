```python
import fastf1
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd

# Carregar os dados da corrida
# Exemplo com o GP de Abu Dhabi 2024 (Substitua por qualquer outra corrida válida)
race = fastf1.get_session(2022, 'Brazil', 'R')  # 'R' = Corrida
race.load()

# Agora você tem todos os dados da sessão carregados
laps = race.laps  # Dados das voltas dos pilotos
laps.head()  # Ver as primeiras linhas para verificar os dados

# Escolher um piloto
hamilton_laps = laps[laps['Driver'] == 'HAM']  # 'HAM' para Lewis Hamilton
lap_data = hamilton_laps.iloc[10]  # Seleciona a primeira volta

# Carregar dados detalhados da telemetria da volta
# The get_telemetry method is now accessed directly from the Lap object
telemetry = lap_data.get_telemetry()

# Vamos exibir alguns dos dados
telemetry[['X', 'Y', 'Speed']].head(30)
```

```python
import plotly.express as px

# Plotando a posição na pista (gráfico interativo com Plotly)
fig = px.scatter(
    x=telemetry['X'],
    y=telemetry['Y'],
    color=telemetry['Speed'],
    title="Posição na Pista - Velocidade de Hamilton",
    labels={"X": "Posição X", "Y": "Posição Y", "color": "Velocidade"}
)

# Ajusta a proporção dos eixos para manter o formato correto do circuito
fig.update_layout(
    yaxis=dict(scaleanchor="x", scaleratio=1),  # Força a mesma escala nos eixos X e Y
    title="Posição na Pista - Velocidade de Hamilton",
    xaxis_title="Posição X",
    yaxis_title="Posição Y"
)

fig.show()

# Gráfico de Velocidade (não precisa de ajustes de proporção)
fig2 = px.line(telemetry, x='Time', y='Speed', title="Velocidade ao longo da Volta - Hamilton")
fig2.show()
```
